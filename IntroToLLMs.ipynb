**What is a large language model(LLM)?**

LLM is a neural network designed to understand generate and respond to human like texts.Consider example of chatgpt , it is a LLM which can understand , generate and respond to human like texts.

At the core of LLM , we have Deep neural network trained on massive amounts of text data.

**Why to say it Large Language Models?**

LLM model has billions of parameters. GPT3 small  had 125 million parameters , GPT3 Medium had 350 million parametrs.

These models do a wide range of NLP tasks :question answering , translation , sentiment analysis and much more.

**LLM VS Earlier NLP models**

LLM models can do wide range of NLP tasks.Earlier language models could not write an email from custom instructions , a task that is trivial for modern LLM's.

NLP models was designed for specific tasks like language translation.

**What makes LLMs so good?**

LLMs are based on the Transformer Architecture which makes LLMs capable  to executing the complex NLP tasks.

**LLM vs GenAI vs DeepLearning vs Machine Learning**

Broadest umbrella is AI and smallest umbrella is LLMs

Ai --> ML --> DL --> LLMs

LLMs represents a specific application of deep learning techniques , leveraging their ability to process and generate human like texts.

GenAI uses deep neural network to create new content such as texts,images and various forms of media.

**Application of LLMs-**

* Content creation like Poem
* Chatbots or virtual assistance
* Machine Translations like English to Hindi language.
* Noval Text Generation
