**Zero Shot , One Shot and  Few Shot Learning-**


**Zero shot learning** refers to the models ability to perform a task without having seen any examples of that task or class during training.It leverages general world knowledge and task instructions to make predictions.INstead of learning from labelled examples , the model relies on descriptive prompts,tasks definations, or semantic similarities.No fine tunning is done for the specific task.Prompt in the zero shot learning contains natural language instructions.Learning style is prompt understanding and world knowledge.

**One Shot Learning-**  refers to learning a new concept from only one labeled example.The model sees just one example of a class or task and is expected to geeralize from it.

**Few Shot LEarning** refers to a models ability to perform a task after being shown only a few (typically 1-100) examples of that task.No fine tunning,just examples in the prompts.Prompts in few shot learning contains prompts along with a few labeled exmples.Learning style is prompt along with pattern recognition from few examples.


In GPT4 model,it is a zero shot learner and few shot learner but if we could give few examples as to how do we expect the result from GPT4 model,it would have been better for GPT4 model to give more concise and accurate result.

* GPT architecture- GPT architecture is only consist of encoder part.GPT3 is a scaled up version of original transformer model,implemented on a larger dataset.

* GPT-3 is an autoregressive, Transformer-based language model trained using self-supervised learning. At each step, it performs a multiclass classification task to predict the next token.

* With such training , GPT3 can do a wide range of other taks like translation,spelling correction etc.

* We dont collect labels for training data,but use the structure of the data itself. Next word in sentence is used as label.And thats why we call GPT3 model as the AutoRegressive model.

* Compared to original transformer architecture,GPT architecture is simpler because there is no encoder part. We just have the decoder.

* Original transformer architecture is consist of 6 encoder and 6 decoder blocks.But GPT3 model is consist of 96 transformer layers with 17 billion parameters.

**How next word prediction happens?**

At 1st iteration we have input as "This".At 2nd iteration we have input as "This is" . At 3rd iteration we have input as "This is an".Such continuous inputs we have in each iterations. Output of previous iteration is used as the input to the next iteration and thats why GPT model is autoregressive in nature.
