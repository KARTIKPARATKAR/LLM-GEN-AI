**Stages Of Building LLMs-**

For creating a LLM , we have to first pretrain the model on large and diverse dataset and secondly we have to fine tune the LLM on specific limited dataset.

LLM Creation = Pre-Training + Fine-Tunning

Let you own a tution classes and have a requirement of creating the MCQ for test of student.SO you can make use of some pretrained LLM model and fine tune this LLM model on the specific dataset of book. So , the content of the book will be used as the specific data for fine tunning. As a result , our model will work better on the specific task of generation of MCQ on any topic available in the book.

Pretrained LLM models are also called foundational LLM models.

**Steps for building a LLM**


1) Train on a large corpus of text data(raw text). Raw text means regular text without any labelling information.

2) First training stage of LLM is also called pretraining. Our aim here is to create an initial pretrained LLM which is base or foundational models. GPT3 model is a pretrained model which is capable of text comletion.

3)After obtaining the pretrained LLM , we can further train LLM on Labelled data which is also known as fine-tunning.

4)THere are 2 popular categories of finetunning-

a)Instruction Finetunning-Labelled dataset consists of instruction-answer pairs eg-text translation , airline customer support.

b)Finetunning for classification tasks-Labelled dataset consist of text and associated label. Eg- Email - Spam or Not-Spam.
