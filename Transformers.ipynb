**Basic Intro to Transformers-**

- Most modern LLM rely on the transformer architecture . Transformer is a deep neural network architecture introduced in research paper called "Attention Is All You Need"

- Original transformers/original research paper was developed for machine translation tasks like conversion of english text to german text.

**Tokenization** - We have english text sentences. We cant give input as whole sentence.We need to break the sentence into words. The process of breaking down of sentence into words and assigning an id or numbers to each word is known as Tokenization.

1.jpg

**Vector Embeddings** -

We have a word tokens along with the token id. We have to tell model that which words are related to each other and by how much value it is related to each other. We will convert these token of words into vectors which represents the relation of one word with other word in the input sentence.

1.jpg



By converting the input text word embeddings into vector representation , we would be able to capture the semantic meaning of the words present in the data.

Job of Decoder architecture is to give the final output.Decoder gives output word one by one in each time step.

**Transfomer Architecture-**

Transformer architecture is consist of two parts. One is encoder and and another is decoder.Encoder converts input text into vectors.Decoder generates output text from encoded vectors.

**Self-Attention Mechanism-**

* Key part of transformers is that , it allows models to weigh importance of different words/tokens relative to each other.

* Self attention enables model to capture long range dependencies.

**Variations of Transformer Architecture-**

* BERT-Bidirectional Encoder Representation from Transformers. BERT predicts hidden words in a sentence. BERT is an encoder architecture.BERT works better in case of sentiment analysis because BERT can see the input data in both the directions.

1.jpg


* GPT-Generative Pretrained Transfomers.GPT generates new words.GPT is a decoder model.
